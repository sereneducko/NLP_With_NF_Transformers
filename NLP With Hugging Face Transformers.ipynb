```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "What Donald Trump said in the rally is frankly radical",
    candidate_labels=["far left", "politics", "far right"],
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "The bulls are rallying in the market today, though economist say that this is a bearish season",
    candidate_labels=['negative', 'up', 'down'],
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "Autobots roll out, the Decepticon are destroying New York City",
    candidate_labels=['terrorism', 'Transformer', 'plane'],
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("But it is said only the <mask> can restore balance to the world.", top_k=2)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("Percy Jackson was born in New York City")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Monster Hunter adalah game yang didevelop oleh Capcom, dimana pemain main sebagai Hunter yang melawan monster dan menggunakan bagian tubuh mereka untuk membuat armor dan senjata."
question = "Apa itu monster hunter?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Pohon Pinus adalah salah satu jenis pohon yang tidak berdaun"
question = "Pohon pinus adalah?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("The jokes was great, but the crowd didn't took it well")
```

```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-id-en")

text_to_translate = "Senja begitu indah"
result = translator(text_to_translate)

print(result[0]['translation_text'])
```

Menurut saya zero - shot classification ini sangat menarik untuk melihat dalam aksi. Karena seakan - akan macam ada ChatBot at home.

Untuk zero shot terlihat guna karena bisa mengetahui makna atau arti kedua dalam sebuah kalimat. Selain itu juga bisa digunakan untuk menkategorikan konten yang diberikan pada feed seseorang dalam sosial media. Tetapi terkadang ada label yang sangat mendominasi padahal tidak ada hubungannya.

Untuk aspek text generation masih agak kurang dalam pengalaman pemakaian. Tetapi mungkin karena kalimat yang diberikan masih sangat singkat, jadi tidak mendapatkan konteks yang banyak. Sama hal seperti answering question, hanya diberikan jawaban yang singkat, dan terkadang tidak make sense

Untuk fill-mask juga sepertinya masih kurang konteks yang diberikan. Tetapi ini bisa digunakan untuk suggestion atau typo correction mungkin.

NER ini sepertinya bisa berguna bagi model untuk memahami kalimat dan struktur bahasa.

Sentiment Analysis dapat digunakan untuk mengevaluasi review atau penilaian dari sebuah restaurant atau service dengan cepat, sehingga mengetahui mana yang negative dengan cepat, agar bisa direview dengan lebih seksama.

Untuk Translation tentu sangat menarik karena bisa digunakan untuk translate bahasa yang asing jadi bahasa yang kita ketahui. Ini bisa bermanfaat dalam skenario diplomat atau hanya untuk menerjemahkan media hiburan favorit.
